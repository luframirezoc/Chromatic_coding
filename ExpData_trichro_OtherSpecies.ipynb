```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spo
import warnings
import scipy.stats
```

```python
def Sens_curves(N_system):
        
    N_points = 25
    l_array = np.loadtxt("Data/F"+str(N_system)+"_opsin1.txt")[0,:]
    Red_mean = np.loadtxt("Data/F"+str(N_system)+"_opsin1.txt")[1,:]
    Green_mean = np.loadtxt("Data/F"+str(N_system)+"_opsin2.txt")[1,:]
    Blue_mean = np.loadtxt("Data/F"+str(N_system)+"_opsin3.txt")[1,:]

    return l_array, (Red_mean), (Green_mean), (Blue_mean)
```

```python
def cost_function_trichromatic_fixedpoints(params):
    hr, hg = params[0], params[1]
    beta_ex = 1
    beta_inh = 1
    return (-hr + Ir + Wrh*np.tanh( beta_inh*( Whr*np.tanh(beta_ex*hr) + Whg*np.tanh(beta_ex*hg) + (Whr+Whg) ) ) + Wrg*np.tanh(beta_ex*hg) + (Wrh + Wrg))**2 + (-hg + Ig + Wgh*np.tanh( beta_inh*( Whr*np.tanh(beta_ex*hr) + Whg*np.tanh(beta_ex*hg) + (Whr+Whg) ) ) + Wgr*np.tanh(beta_ex*hr) + (Wgh + Wgr))**2 
```
